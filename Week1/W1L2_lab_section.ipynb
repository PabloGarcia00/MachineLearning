### Fitting a simple linear-regression model
To start with linear regression, we will use the function `smf.ols()`, where ols is short for *ordinary least squares*, referring to the method to fit the regression line to the data. 

The basic syntax is:

``smf.ols(y ~ x, data=df).fit()``,

where `y` is the response, `x` is the predictor, and `df` is the pandas data frame in which these two variables are kept. Note that the first part of the code initializes the model object in `Python`, and the `.fit()` call performs the actual model fit.

In the code below, you can see how we can fit a simple linear regression model, with `medv` as the response and `lstat`  as the predictor. 

```python
lm_fit = smf.ols('medv ~ lstat', data=boston).fit()
```

We can get information about the fitted model, using `lm_fit.summary()`. This gives us $p$-values and standard errors for the coefficients, as well as the $R^2$ statistic and $F$-statistic for the model (and more).


```python
print(lm_fit.summary())
```

There is a lot of output of `lm_fit.summary()`. For now, focus on the following:
* The model: $y = \beta_0 + \beta_1 x$ has the **intercept** listed in the row `Intercept` and the **slope** in `lstat`. So we can see that the model estimates the coefficient $\beta_0=34.6$  and $\beta_1-0.95$.
* the **t-statistics on the intercept** is 61.4 with a very low associated p-value (`P>|t|`), indicating that the intercept is significantly different from 0. 
* the **t-statistics on the slope* is -24.5 with a very low associated p-value, indicating that the slope is significantly different from 0, which means that there is **a substantial association between predictor (lstat) and response (medv)**.
* The $R^2$ value of the fitted linear model is 0.544
* The adjusted $R^2$ is 0.543
* The F-statistics is 601.6, with a very low associated p-value, which indicates that there is **at least one predictor significantly related to the response**
* The 95% confidence interval on the coefficient for lstat is from -1.026 (2.5%) to -0.874 (97.5%)

### Plotting the simple linear-regression model
We define a function `plot_data_and_model` that will plot the `medv` and `lstat` data along with the fitted regression line. The function creates a scatter plot using `sns.scatterplot`, and we add the model fit using `lm.predict()` and `plt.plot`. 

```python
## This function plots an abline for the model y = ax + b in the interval [x_min,x_max]
def plot_abline(a, b, x_min, x_max, color = 'r', linestyle = '-'):
  plt.plot([x_min, x_max], [a*x_min+b, a*x_max+b], color=color, linestyle=linestyle)
    

# Function to plot the prediction of the model and the data
def plot_data_and_model(df, lm, dep_var, indep_var, n_points=100, plot_data=True, color_model='r', label='model'):
  if(plot_data):
    sns.scatterplot(x=df[indep_var], y=df[dep_var], label='Data')
  x = pd.DataFrame(np.linspace(df[indep_var].min(), df[indep_var].max(), n_points), columns = [indep_var])
  plt.plot(x, lm.predict(x), color=color_model, label=label)
  plt.legend()
```

### Diagnostic plots

Next we examine some diagnostic plots, several of which were discussed in Section 3.3.3. Four diagnostic plots are automatically
produced with the function `diagnostics_plots()`, which we define below.

Run the code below to define the function.


```python
def diagnostics_plots(lm):
  """
  Plots the four diagnostic plots for lm() output.
  1) Residuals vs Fitted
  2) Normal Q-Q
  3) Scale-Location
  4) Residuals vs Leverage
  """

  fitted_values = lm.fittedvalues
  residuals = lm.resid
  s_residuals = lm.resid_pearson
  #s_residuals = lm.get_influence().resid_studentized_external

  fig, axes = plt.subplots(2, 2, figsize=(10, 10))

  # Plot 1: Residuals vs fitted
  sns.regplot(x=fitted_values, y=residuals, ax=axes[0, 0], marker='.', lowess=True, line_kws={"color": "red"})
  axes[0, 0].set_xlabel("Fitted Values", fontsize=15)
  axes[0, 0].set_ylabel("Residuals", fontsize=15)
  axes[0, 0].set_title('Residuals vs fitted', fontsize=15)
  axes[0, 0].grid()

  # Plot 2: Standardized residuals
  sm.qqplot(residuals, ax=axes[0, 1],fit=True, line="45", color=sns.color_palette()[0], marker='.')
  axes[0, 1].set_xlabel("Theoretical Quantiles", fontsize=15)
  axes[0, 1].set_ylabel("Studentized residuals", fontsize=15)
  axes[0, 1].set_title('Normal Q-Q', fontsize=15)
  axes[0, 1].grid()

  # Plot 3: Scale-Location
#  sns.regplot(x=fitted_values, y = np.sqrt(s_residuals), ax=axes[1, 0], marker='.', lowess=True, line_kws={"color": "red"})
  sns.regplot(x=fitted_values, y = s_residuals, ax=axes[1, 0], marker='.', lowess=True, line_kws={"color": "red"})
  axes[1, 0].set_xlabel('Fitted values', fontsize=15)
#  axes[1, 0].set_ylabel('sqrt( Studentized residuals )', fontsize=15)
  axes[1, 0].set_ylabel('Studentized residuals', fontsize=15)
  axes[1, 0].set_title('Scale-Location', fontsize=15)
  axes[1, 0].grid()

  # Plot 4: Influence Plot
  sm.graphics.influence_plot(lm, ax=axes[1,1], criterionstr='Cooks', size = 1)
  axes[1, 1].set_xlabel('h leverage', fontsize=15)
  axes[1, 1].set_ylabel('Studentized residuals', fontsize=15)
  axes[1, 1].set_title('Residuals vs leverage', fontsize=15)
  axes[1,1].grid()

  plt.tight_layout()
  plt.show()
```

We can now show the plots for the fitted linear-regression model:

```python
diagnostics_plots(lm_fit)
```

### High leverage

A point with a high **leverage** value has an unusual x observation.  In other words, the observed value of a predictor is very unusual compared to other values. Leverage statistic $h$ can be used to spot high leverage values.

$h_i = \frac{1}{n} + \frac{(x_i - \bar{x})^2}{\sum_{j=1}^n (x_j-\bar{x})^2}$

The leverage can be calculated for all samples. This can be obtained using:

`lm_fit.get_influence().hat_matrix_diag`.

The average leverage is $\frac{\sum_{i=1}^n h_{i}}{n} = \frac{p+1}{n}$. To find high leverage points, a common used cut-off value is 3 times the average. This means that we mark data points as high leverage if: 

$h_{i} > 3\cdot \frac{p+1}{n}$.

With the code below, you can make a plot to show leverage vs studentized residuals. The dasked black line shows the h cut-off for the leverage.

**Exercise:**
* Make the plot using the code below.
* Are there points with high leverage?

```python
leverage = lm_fit.get_influence().hat_matrix_diag
p = lm_fit.df_model
n = len(boston)
h_cut_off = 3*(p+1)/n

sns.regplot(x=leverage, y=s_residuals, marker='.', lowess=True, line_kws={"color": "red"})
plt.plot((h_cut_off, h_cut_off), (s_residuals.min(), s_residuals.max()), color='black', linestyle='--')
plt.xlabel('h leverage')
plt.ylabel('Studentized residuals')
plt.show()
```

### Cook's distance
Based on the studentized residuals, outliers in the predicted value (y) can be identified. Using h-leverage, outliers in de predictors (x) can be found. With **Cook's distance**, **influential points** can be found. An influential point is a point that creates a large change in the fitted linear model if it is removed from the data set. It is calculated based on the residuals and the leverage:

$D_i = \frac{r_i^2}{p\cdot \textrm{MSE}} \cdot \frac{h_{ii}}{(1-h_{ii})^2}$,

where $r_i$ is the residual for point $i$, $p$ is the number of predictors in the model, $\textrm{MSE}$ is the mean squared error, and $h_{ii}$ is the leverage of point $i$.

The Cook's distance can be obtained from the data using `lm_fit.get_influence().cooks_distance[0]`. A commonly used cut-off for the Cook's distance is three times the average Cook's distance.

$D_i > \bar{D}/3$

The code below identifies the outlier points based on Cook's distance and identifies these in the plot

```python
cooks_d = lm_fit.get_influence().cooks_distance[0]
c_cut_off = cooks_d.mean()*3

sns.regplot(x=leverage, y=s_residuals, marker='.', lowess=True, line_kws={"color": "red"})
plt.plot(leverage[cooks_d>c_cut_off], s_residuals[cooks_d>c_cut_off], 'r*')

plt.xlabel('h leverage')
plt.ylabel('Studentized residuals')
plt.show()
```

And we can again find the samples in the dataset that have a Cook's distance larger than the threshold.

### Using all predictors 

If we want to use all 12 predictos in the `Boston` dataset, we should use the formula:<br>
`medv ~ crim + zn + indus + chas + nox + rm + age + dis + rad + tax + ptratio + lstat`

As it is a bit cumbersome to type this in, we will use a helper function to create a formula using all predictors

`lm_formula(df, dependent_var, excluded_cols=[])`

Where `df` is the dataframe, `dependent_var` is the variable name that we use as repsonse, and `excluded_cols` is an optional list of column names that you want to exclude. If you leave the latter empty, you will get the formula to predict the dependent variable from all the other variables in the data frame.


```python
def lm_formula(df, dependent_var, excluded_cols=[]):
  '''
  Generates the R style formula for statsmodels (patsy) given
  the dataframe, dependent variable and optional excluded columns
  as strings
  '''
  df_columns = list(df.columns.values)
  df_columns.remove(dependent_var)
  for col in excluded_cols:
    df_columns.remove(col)
  return dependent_var + ' ~ ' + ' + '.join(df_columns)
```

### The variance inflation factor

When we use multiple predictors, there is always a change of **colliniarity**, which is the situation that two or more predictors are closely related to each other. This can influence the quality of the fitted linear model, and it is adviced to identify the collinearity.

This can be done by calculating the **variance inflation factor**, $\textrm{VIF}(\hat{\beta_j})$:

$\textrm{VIF}(\hat{\beta_j}) = \frac{1}{1-R_{X_j|X_{-j}}^2}$,

Where $R_{X_j|X_{-j}}^2$ is the $R^2$ from a regression of $X_j$ onto all the other predictors, which can be interpreted as a value indicating how wel predictor $j$ can be predicted from the other predictors. If this is very high (close to 1), the predictor is collinear to other predictors and we get a very high $\textrm{VIF}(\hat{\beta_j})$. If predictr $j$ cannot be predicted from the other predictors (and therefore contains new information), $R_{X_j|X_{-j}}^2$ is zero and $\textrm{VIF}(\hat{\beta_j})$ will be 1.0.

As a rule of thumb, a cut-off of 5 or 10 is often used to indicate predictors with problematic amount of collinearity:

$\textrm{VIF}(\hat{\beta_j})>5$ or $\textrm{VIF}(\hat{\beta_j})>10$

We define the function `vif()` to compute variance
inflation factors. 

```python
# Define a function to calculate the variance inflation factors for all predictors
from statsmodels.stats.outliers_influence import variance_inflation_factor
def vif(df, dep_var):
  df_no_dep_var = df.drop(dep_var, axis=1)
  exog_df =  sm.add_constant(df_no_dep_var.values)
  vifs = pd.Series( [variance_inflation_factor(exog_df, i) for i in range(1,exog_df.shape[1])  ],
                    index=df_no_dep_var.columns, name = 'vif')
  return(vifs)

```

### Comparing two models using ANOVA

We can compare two models using the function `sm.stats.anova_lm()` function. We can use this to quantify the extent to which the quadratic fit is superior to the linear fit:

```python
mlm_fit_1 = smf.ols(formula='medv ~ lstat', data=boston).fit()
mlm_fit_2 = smf.ols(formula='medv ~ lstat + I(lstat**2)', data=boston).fit()
sm.stats.anova_lm(mlm_fit_1, mlm_fit_2)
```

Here, `mlm_fit_1` represents the linear submodel containing only one predictor, 'lstat', while `mlm_fit_2` is the quadratic model that has two predictors, 'lstat' and 'lstat^2'.

The `anova_lm()` function performs a hypothesis test
comparing the two models. The  null hypothesis is that the two models fit the data equally well, and the alternative hypothesis is that the second model is superior. 

Here the $F$-statistic is $135$ and the associated $p$-value is virtually zero. This provides very clear evidence that the model containing the quadratic term is far superior to the model that only contains the predictor 'lstat'. This is not surprising, since earlier we saw evidence for non-linearity in the relationship between `medv` and `lstat`.

## Adding polynomial terms

We might want to try to add higher-order polynomial terms. This can be done using `medv ~ lstat + I(lstat**2) + I(lstat**3)`, etc. To assist us in that process, we define a function `poly_formula()` that can create the formula for any given polynomial order. 

```python
# Function to make a formula to add polynomial terms
def poly_formula(dep_var, indep_var, order: int) -> str: 
  assert order >= 1, "poly_formula can only be used for order >= 1"
  formula = dep_var + ' ~ ' + indep_var + ' + ' + ' + '.join('I(%s**%d)'%(indep_var,i) for i in range(2,order+1))
  return(formula)
```

For instance, to predict 'medv' based on a 5th polynomila model of 'lstat', we can call `poly_formula('medv', 'lstat', 5)`.

The code below uses this 5th-order polynomial to fit a linear model and shows the results.

**Exercise:**

* Look at the summary. Are all terms significantly related to the response value?
* Change the code to fit a 7-order polynomial. Does it make sense to add these high-order terms?

```python
# Fit a 5-th order polynomial model
order = 5
mlm_fit_poly = smf.ols(formula=poly_formula('medv', 'lstat', order), data=boston).fit()

# Print the results
print(mlm_fit_poly.summary())

# Plot the model predictions with the model on top of the data
plt.figure(figsize=(8,6))
plot_data_and_model(boston, mlm_fit_poly, 'medv', 'lstat', color_model='r', label = 'poly model')
plt.legend()
plt.grid()
plt.show()
```

This suggests that including additional  polynomial terms, up to fifth order, leads to an improvement in the model fit! However, further investigation of the data reveals that no polynomial terms beyond fifth order have significant $p$-values
in a regression fit.
