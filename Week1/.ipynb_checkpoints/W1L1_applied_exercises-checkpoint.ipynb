v. Use the `sns.histplot()` function in combination with `plt.subplots()` to produce six histograms with differing numbers of bins for a few of the quantitative variables and put the plots in a two-row, three-column structure.

````python
import numpy as np
def indices_array_generic(m,n):
    r0 = np.arange(m) # Or r0,r1 = np.ogrid[:m,:n], out[:,:,0] = r0
    r1 = np.arange(n)
    out = np.empty((m,n,2),dtype=int)
    out[:,:,0] = r0[:,None]
    out[:,:,1] = r1
    return out
````

````python
arr = indices_array_generic(2,3).reshape(6,2)
````

````python
for n in arr:
    x,y = n
    print(x)
````

````python
fig, axes = plt.subplots(2, 3, figsize=(6, 9))
bins = [3, 6, 9, 12, 15, 18]
df_quant = df.select_dtypes(include="number")
i =0
for p in arr:
    x, y = p
    sns.histplot(df_quant, x=df_quant.columns[i], bins=bins[i], ax=axes[x,y])
    i +=1
````
