Grouping objects based on similarities without labels. 

**Principal Component Analysis**

summarize a set of variable using linear combinations. 

The principale component of a set of features $X_1, X_2,... X_p$ is a linear combination of these features where the variance is maximized on $Z_i$: 
$$\Large Z_{1}=\phi_{11}X_{1}+\phi_{21}X_{2}+\cdot\cdot\cdot+\phi_{p1}X_{p}$$

The term $\large \phi$ here is a normalized (column mean of the variable vector is zero) weight/loading/coefficient of the principal component, by normalized is meant that the sum of $\large \phi^2$ is one, this is necessary as the variance would otherwise get arbitrarely large. The new variable $\large Z_i$ is derived from a linear combination of the old variables.
The matrices that are returned are the eigenvalues which contains the samples and their coordinates in PC space, the loadings/weight/coefficient which are the values that correspond to each of the variables, and the sigma matrix/eigenvector which is the variance of each of the componenents. 

*Multiple principal components* 

The second principal component must be orthogona/uncorrelated to the other prinicpal components as it should describe other properties of the data. This can be visualized in a biplot using the 2 PCs, additionally we can plot the loadings of the variables as we have their loadings on PC1 and 2. 




*In practice PCA also fits a hyperplane to the points in n-dimensional space only here there is no y-label as in partial least squares.*

**Proportion of variance explained**

is used to select the number of PCs to include to explain most of the variance in the data, we visualize this in a scree plot where we plot the eigenvector values and look for an elbow. 


**non-linear data**

When data is non-linear PCA has trouble with capturing the essence of the data while reducing the dimensions. To remedy this kernel PCA can be used. This method maps data into higher dimensional space, where data does become linear.

Another method is Independent Component Analysis (ICA). Where PCA is build on finding correlation and maximizing variance, ICA uses linear combinations to create new features which have mutual information is zero. It assumes the data is non-gaussian and independent. This method is used often with EEG data where blink spikes are removed by first reducing complexity using PCA followed by ICA to seperate the blink signal, as PCA allows for reconstruction of the data removing the components that constitute the blink spike allows for cleaning the signal. 

**K-means clustering**

Prespecifying the number of clusters K, and seperating the data into these K subgroups. 

* we split the data up into K-subgroups that are non-overlapping
* Clusters are good if the intercluster(among) distance/variation is big and the intracluster (within) variation is small
* We thus aim to optimize that we assing the n-datapoints to its cluster so that the within cluster variation is the smallest. To assign datapoints we can use different distance (dissimilarity) methods.

The algorithm works as follows:

1. assign each datapoint randomly to a cluster.
2. calculate the centroid of the clusters.
3. assign each datapoint to its closest centroid
4. recalculate the centroids

Doing this iterative calculation of the centroids and relocation of the centroid the data is partitioned better and better. This also highlights a drawback of finding a "local minimum" depending on its starting point, to remedy this we can start the algorithm from different places in the hopes to find the "global minimum". 



**Hierarchical clustering**

Stepwise groups objects together in a hierarchical manner, this data is visualized using dendrograms. 

Algorithm:

1. Group the two points are closest together in the dissimilarity matrix.
2. Calculate the distance between the fused cluster, different methods: single, complete average, and centroid.
3. do step one again until we grouped all the points together.

Distance methods 

* single link &rarr; the smallest dissimilarity (nearest neighbour)
* complete link &rarr; the largest dissimilarity (farthest neighbour)
* average link &rarr; the average distance of all points to the point of interest
* centroid link &rarr; the distance to the centroid of the cluster(first calculate the average coords in n-dimensional space).

Dissimilarity methods:

* euclidean distance &rarr; very commonly used.
* correlation &rarr; commonly used when features are measurements across time.
