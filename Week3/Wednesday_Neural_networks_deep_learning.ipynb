# Wednesday: Neural networks and deep learning

Neuron: fires when sufficiently stimulated. 

The output of a neuron is modelled as a weighted linear combination of inputs $X$ plus a bias $\beta_0$. Over this an activation function $f$ is used to make a decision. 

$$\Large y = f(\sum_j\beta_jx_j+\beta_0)$$

The activation function can be of different nature:

* Linear: outputs the same linear combination itself
* Sigmoid: ouputs numbers between 0 and 1
* Step: classifies into two groups
* ReLU: outputs positive numbers

A neural network aims to determine the weights for a feature $x$ given an response $y$, this is similar to linear regression. 

Gradient Descent: The Loss/error function (how many samples were wrongly classified), gradient descent is the process of minimizing this "Loss" by adjusting the parameters $\beta$. Often the path to the minima is found by following the steepest decline using the derivative. The step by which you reduce $\beta$ is called the learning rate $\alpha$.

 


A trick to aid in the problem of finding the optimal minima given a certain learning rate $\alpha$ is to add a momentum term. This adds so called "intertia" which aids in finding the minima as seen in the right-most image above, by adding a term in the previous direction whereby infinite oscillation around the minima is stopped.

**Convolutional Neural Networks (CNNs)**

Especially succesful is written letter recognition. This method is computationally heavy and takes a lot of data to be trained. 

The weights $\beta$ are not numbers but convolutional filters i.e. it looks at a small neighbourhood of pictures and applies an operation to this. The neighbourhood is captured by a "kernel", this kernel slides over the picture space. 

These kernels generate many feature maps, The fact that the same kernel is used all over the image reduces the total computation and is partly the reason why CNNs were succesful.

The proces of pooling shrinks the image after each feature map that way the model will eventually see the whole image as it will become as big as the kernel itself.
