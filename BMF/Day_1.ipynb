```python
import pandas as pd 
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt 
import statsmodels.api as sm
import statsmodels.formula.api as smf
from sklearn.model_selection import train_test_split
from sklearn.metrics import confusion_matrix
```

```python
# load in the data

df = pd.read_csv("../../ml2023/data_labeled_fixed.csv")
df.columns = df.columns.str.replace("-", "")
samples = df[["CELLTYPE", "CLASS"]]
exp = df.drop(df.columns[-2], axis = 1)
exp.head()
```

```python
def formula(df, independent, exclude=[]):
    cols = df.columns.values.tolist()
    cols.remove(independent)
    cols = [x for x in cols if x not in exclude]
    return independent + " ~ " + " + ".join(cols)
```

## Regression 

### Calm1 regression 

```python
### multiple linear regression using only rods that are bipolar
rodbip = exp.loc[exp["CLASS"] == "Rod Bipolar",].iloc[:,:-1]
#cols = rodbip.columns.tolist()
#cols.remove("Calm1")

# split the data 
rodbip_train, rodbip_test = train_test_split(rodbip, random_state=1)
#X_train, Y_train = rodbip.loc[:, cols], rodbip["Calm1"]

# define the regression formula 
calm1_formula = formula(rodbip_train, "Calm1")

calm1_rodbip = smf.ols(calm1_formula, rodbip_train).fit()
calm1_rodbip.summary()
```

### logistic regression 

```python
def logmodel(class_type, train_data, test_data):
    print(10*"#"+class_type+10*"#" + "\n")
    
    train_data = train_data.copy()
    bool_test = (test_data.CLASS == class_type)
    train_data.CLASS = (train_data.CLASS == class_type).astype(int)
    
    tmp_formula = formula(train_data, "CLASS")
    model = smf.logit(tmp_formula, train_data).fit()
    
    p = model.predict(test_data) > 0.5
    conf_matrix = confusion_matrix(bool_test, p, normalize="all")
    print(model.summary())
    print("\n" + "# Confusion Matrix")
    print("\n", pd.DataFrame(conf_matrix, index=["True", "False"], columns=["Pred_True", "Pred_False"]), "\n\n")
    
    return model
```

```python
for class_type in class_types_all:
    train_data, test_data = train_test_split(high_all, random_state=1)
    logmodel(class_type, train_data, test_data)
```
